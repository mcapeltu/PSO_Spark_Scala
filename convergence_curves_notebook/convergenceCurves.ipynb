```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data = pd.read_csv('convergence_curve.csv', header=None)
```

```python
fitness_values=data[0].values
```

```python
plt.plot(fitness_values, label='Best Fitness Value')
plt.xlabel('Iteration')
plt.ylabel('Fitness Value')
plt.title('Convergence Curve')
plt.legend()
plt.grid(True)
plt.show()
```
